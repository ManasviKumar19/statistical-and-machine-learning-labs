```python
data = pd.read_csv("C:/Users/manas/Downloads/data_alexithymia_csv.csv")
```

```python
X_total_data = data.drop(['Sex','CESD'],axis=1)
```

```python
X_total_data
```

```python
Y_full_data = data['CESD']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_total_data, Y_full_data, test_size=0.2, random_state=123)
```

### Regression object

```python
pls = PLSRegression(n_components=5)
```

### Fitting the Data


```python
pls.fit(X_train, y_train)
```

```python
y_pred = pls.predict(X_test)
```

### Getting the MSE

```python
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```
